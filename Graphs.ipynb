```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(".\Graphs.csv", names=['epochs','D1','D2','D3','D4','G1','G2','G3'])
df.head()
epochs=df['epochs']
D1=df['D1']
D2=df['D2']
D3=df['D3']
D4=df['D4']

G1=df['G1']
G2=df['G2']
G3=df['G3']
print len(D1)
print D1.shape
```

```python
fig = plt.figure(figsize=(18,10))
fig1 = fig.add_subplot(2,4,1)
plt.plot(D1) 
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('D_loss_real', fontsize=12)

fig2 = fig.add_subplot(2,4,2)
plt.plot(D2)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('D_loss_fake', fontsize=12)

fig3 = fig.add_subplot(2,4,3)
plt.plot(D3)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('D_loss_cls', fontsize=12)

fig4 = fig.add_subplot(2,4,4)
plt.plot(D4)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('D_loss_gp', fontsize=12)

fig5 = fig.add_subplot(2,4,5)
plt.plot(G1)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('G_loss_fake', fontsize=12)

fig6 = fig.add_subplot(2,4,6)
plt.plot(G2)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('G_loss_rec', fontsize=12)

fig7 = fig.add_subplot(2,4,7)
plt.plot(G3)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('G_loss_cls', fontsize=12)


fig8 = fig.add_subplot(2,4,8)
plt.plot(D3, color='k')
plt.plot(G3, color='g')
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('Classification Loss', fontsize=12)
fig8.legend(labels=['D_cls_loss','G_cls_loss'])
plt.tight_layout()
plt.show()
```
